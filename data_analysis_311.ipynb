## Tabulate functions

```python
def tabulate(attribute_set, new_service_requests):
  for attr in attribute_set:
    
    freq_table = new_service_requests[attr].value_counts()
    percentages = new_service_requests[attr].value_counts(normalize=True) * 100
    newdf = pd.DataFrame({'Counts' : freq_table, 'Percentages': percentages})
    
    total_count = sum(newdf['Counts'])
    display(newdf)
    
    print(f'Total instances: {total_count}')
    

def tab_single(attribute, table):
  freq_table = table[attribute].value_counts()
  percentages = table[attribute].value_counts(normalize=True) * 100
  newdf = pd.DataFrame({'Counts': freq_table, 'Percentages' : percentages})
  total_count = sum(newdf['Counts'])
  display(newdf)
    
  print(f'Total instances: {total_count}')

# Uncomment to tabulate service_request data
# tabulate(attribute_set, new_service_requests)
```

```python

# Encoding libraries
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Algorithm libraries
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB


# Statistical Data 
from sklearn.metrics import classification_report


categorical_features = [col for col in raw_data.columns if raw_data[col].dtype == 'object' and col != 'category']
print(categorical_features)

one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")

X = raw_data.drop('category', axis=1)
y = raw_data['category']

transformed_X = transformer.fit_transform(X)
print(transformed_X.shape)
transformed_X = pd.DataFrame(transformed_X)
# transformed_X.head()

dummies = pd.get_dummies(raw_data[categorical_features])
display(dummies)

empty_transformed = pd.get_dummies(parsed_empty[categorical_features])
display(empty_transformed)
```

```python
from pandas.plotting import scatter_matrix
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Split dataset into test and train
array = raw_data['category']
X_train, X_test, Y_train, Y_test = train_test_split(dummies, array, test_size=0.3, random_state=1)

# Algorithms going to be tested
models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))

# evaluate each model
results = []
algorithms = []

# Use 10-fold cross validation to estimate the accuracy of each model
for algorithm, model in models:
    k_fold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True)
    cv_estimates = cross_val_score(model, X_train, Y_train, cv=k_fold, scoring='accuracy', )
    results.append(cv_estimates)
    algorithms.append(algorithm)
    print('%s: %f ' % (algorithm, cv_estimates.mean()))


# Compare algorithms
plt.boxplot(results, labels=algorithms)
plt.title('Algorithm Comparison')
plt.show()

for algorithm, model in models:
    #Make predictions on test dataset
    model.fit(X_train, Y_train)
    model_predictions = model.predict(X_test)
         
    # Evaluate predictions
    print("\nAccuracy Score: ", accuracy_score(Y_test, model_predictions))
    print("\nClassification Report: \n", classification_report(Y_test, model_predictions))
```

```python

coordinate_data = service_requests['location_1']
        
parsed_coordinates = []
        
for point in coordinate_data:
    if isinstance(point,float):
        continue
    
    parsed_coordinates.append((float(point.split()[1][1:]), float(point.split()[2][:-1])))
    
    
df = pd.DataFrame(parsed_coordinates, columns=['Longitude', 'Latitude'])


USA = ((-124.72,  -66.95), (23.55, 50.06))

from datashader.utils import lnglat_to_meters as webm
x_range,y_range = [list(r) for r in webm(*USA)]

plot_width  = int(2000)
plot_height = int(plot_width*7.0/12)

background = "black"

from functools import partial
from datashader.utils import export_image
from datashader.colors import colormap_select, Greys9

export = partial(export_image, background = background, export_path="export")
cm = partial(colormap_select, reverse=(background!="black"))

cvs = ds.Canvas(plot_width, plot_height, *webm(*USA))
agg = cvs.points(df, 'Longitude', 'Latitude')

export(tf.shade(agg, cmap = cm(Greys9,0.25), how='linear'),"census_gray_linear")
```
