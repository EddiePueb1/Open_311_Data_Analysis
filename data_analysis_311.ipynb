## Tabulate functions

```python
def tabulate(attribute_set, new_service_requests):
  for attr in attribute_set:
    
    freq_table = new_service_requests[attr].value_counts()
    percentages = new_service_requests[attr].value_counts(normalize=True) * 100
    newdf = pd.DataFrame({'Counts' : freq_table, 'Percentages': percentages})
    
    total_count = sum(newdf['Counts'])
    display(newdf)
    
    print(f'Total instances: {total_count}')
    

def tab_single(attribute, table):
  freq_table = table[attribute].value_counts()
  percentages = table[attribute].value_counts(normalize=True) * 100
  newdf = pd.DataFrame({'Counts': freq_table, 'Percentages' : percentages})
  total_count = sum(newdf['Counts'])
  display(newdf)
    
  print(f'Total instances: {total_count}')

# Uncomment to tabulate service_request data
# tabulate(attribute_set, new_service_requests)
```

```python
from pandas.plotting import scatter_matrix
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Split dataset into test and train
array = raw_data['category']
X_train, X_test, Y_train, Y_test = train_test_split(dummies, array, test_size=0.3, random_state=1)

# names going to be tested
models = []
#models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))

# evaluate each model
results = []
names = []

# Use 10-fold cross validation to estimate the accuracy of each model
for algorithm, model in models:
    k_fold = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)
    cv_estimates = cross_val_score(model, X_train, Y_train, cv=k_fold, scoring='accuracy', )
    results.append(cv_estimates)
    names.append(algorithm)
    print('%s: %f ' % (algorithm, cv_estimates.mean()))


# Compare names
plt.boxplot(results, labels=names)
plt.title('Algorithm Comparison')
plt.show()

for algorithm, model in models:
    #Make predictions on test dataset
    model.fit(X_train, Y_train)
    model_predictions = model.predict(X_test)
         
    # Evaluate predictions
  
    print("\nAccuracy Score: ", accuracy_score(Y_test, model_predictions))
    print("\nClassification Report: \n", classification_report(Y_test, model_predictions))
```

# DBSCAN 

```python
from matplotlib import pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn import metrics
from sklearn.cluster import DBSCAN
from collections import Counter
```

```python
coordinate_data = service_requests['location_1']
        
coordinates = []
        
for point in coordinate_data:
    if isinstance(point,float):
        continue
    
    coordinates.append((float(point.split()[1][1:]), float(point.split()[2][:-1])))
    
coordinates = np.array(coordinates)
```

### Plot Coordinates Before Clustering

```python
x, y = coordinates.T
plt.scatter(x,y)
plt.show()
```

### Finding Optimal eps Value

```python
neighbors = NearestNeighbors(n_neighbors=50)
neighbors_fit = neighbors.fit(coordinates)
distances, indices = neighbors_fit.kneighbors(coordinates)

distances = np.sort(distances, axis=0)
distances = distances[:,1]

#Zoom in on curve
plt.xlim(11900, 12100)
plt.ylim(0, 0.0175)

plt.plot(distances)
```

### Run DBSCAN Algorithm

```python
Density_Scan = DBSCAN(eps=0.013, min_samples=50).fit(coordinates)
labels = Density_Scan.labels_

# Number of clusters, ignoring noise
noise_points = list(labels).count(-1)
cluster_amount = len(set(labels)) - (1 if -1 in labels else 0)

print("Estimated number of clusters: %d" % cluster_amount)
print("Estimated number of noise points: %d" % noise_points)
```

```python
cluster_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[Density_Scan.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(cluster_labels))]
for k, col in zip(cluster_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    # Displays points that are core samples
    xy = coordinates[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

    # Displays points that are not core samples 
    xy = coordinates[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=4,
    )

plt.title("DBSCAN Results")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```
