## Data Parsing
#### Retrived data is parsed into a service_request dictionary for easy access

```python
service_requests = dict()
parsed_attributes = dict()
attribute_set = set()
for attr in attributes:
    lowercase = attr.lower()
    lowercase = lowercase.replace(' ', '_')
    # print(f"{attr} : {lowercase}")
    attribute_set.add(lowercase)
    parsed_attributes[attr] = lowercase
    service_requests[lowercase] = new_service_requests[attr]

# parsed_attributes


new_service_requests.rename(columns=parsed_attributes, inplace=True)
display(new_service_requests)
```

```python
def tabulate(attribute_set, new_service_requests):
  for attr in attribute_set:
    
    freq_table = new_service_requests[attr].value_counts()
    percentages = new_service_requests[attr].value_counts(normalize=True) * 100
    newdf = pd.DataFrame({'Counts' : freq_table, 'Percentages': percentages})
    
    total_count = sum(newdf['Counts'])
    display(newdf)
    
    print(f'Total instances: {total_count}')

tabulate(attribute_set, new_service_requests)
```

```python

coordinate_data = service_requests['location_1']
        
parsed_coordinates = []
        
for point in coordinate_data:
    if isinstance(point,float):
        continue
    
    parsed_coordinates.append((float(point.split()[1][1:]), float(point.split()[2][:-1])))
    
    
df = pd.DataFrame(parsed_coordinates, columns=['Longitude', 'Latitude'])


USA = ((-124.72,  -66.95), (23.55, 50.06))

from datashader.utils import lnglat_to_meters as webm
x_range,y_range = [list(r) for r in webm(*USA)]

plot_width  = int(2000)
plot_height = int(plot_width*7.0/12)

background = "black"

from functools import partial
from datashader.utils import export_image
from datashader.colors import colormap_select, Greys9

export = partial(export_image, background = background, export_path="export")
cm = partial(colormap_select, reverse=(background!="black"))

cvs = ds.Canvas(plot_width, plot_height, *webm(*USA))
agg = cvs.points(df, 'Longitude', 'Latitude')

export(tf.shade(agg, cmap = cm(Greys9,0.25), how='linear'),"census_gray_linear")
```
